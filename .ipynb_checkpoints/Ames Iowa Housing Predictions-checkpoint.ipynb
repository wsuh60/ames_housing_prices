## EDA

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor 
from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
train = pd.read_csv('./train.csv')
test = pd.read_csv('./test.csv')

train.head() # used .head this time lol
```

```python
train.shape
```

```python
# Start by getting numeric columns only and dropping some columns that I don't need. 
# Masking with numerical features columns will avoid dealing with true nulls in categorical features.

numerical_features = list(train._get_numeric_data().drop(['Id', 'PID', 'SalePrice'], axis = 1))
target = train['SalePrice']
```

```python
# highlight objects/ category columns:

categorical_features = list(train.dtypes[train.dtypes == 'object'].index)
```

```python
numeric_train = train[numerical_features]
numeric_test = test[numerical_features]

# There are some missing values for numerical features... I'll deal with that later.  *used .isnull().sum()*
```

```python
# Let's check out the kde for numeric_features to see if there's anything interesting going on...

n_features1 = 10
n_rows = 13
fig, axs = plt.subplots(n_rows,3, figsize = (20,60))

for k in axs:
    for i,j in zip(numeric_train, axs.flat):
        g = train[numerical_features][i].plot.kde(ax = j, title = i)
        g.get_xaxis().tick_bottom()
        g.get_yaxis().tick_left()
```

#### Observations for KDE plot for numeric_df:

- A lot of remodeling/additions to homes happened around 2000. 
- A lot of homes were built around 2000.
- Most common Overall condition is 5.
- Most homes have 3 bedrooms above ground.
- Most garages were built in 2000. (and most are for 2 cars).
- Looks like June is the most popular month for homes sold.

```python
# Let's check out categorical data 
# I might have to drop some categorical data that have a lot of missing values but more on that later...

n_features1 = 10
n_rows = 15
categorical_train = train[categorical_features]

fig, axs = plt.subplots(n_rows,3, figsize = (20,60))

fig.subplots_adjust(hspace=.8, wspace=.3)


for k in axs:
    for i,j in zip(train[categorical_features], axs.flat):
        g = sns.countplot(x=i, data=train[categorical_features], ax=j)
        plt.setp(j.get_xticklabels(), rotation=90)
```

#### Interesting Observation for Object_df:

- North Ames had a lot homes built (over 300)!
- Most homes were sold as warranty deed.

### Iteration 1: GridSearch w/o Standard Scale (submission.csv)

```python
# Filled NaN values with a large negative values.

og_numeric_train = train[numerical_features]
og_numeric_test = test[numerical_features]

numeric_train = numeric_train.fillna(-999)
numeric_test = numeric_test.fillna(-999)
```

```python
params = {}
gs = GridSearchCV(RandomForestRegressor(), param_grid= params)
gs.fit(numeric_train, target)
print('Random Forest GridSearch best score:', gs.best_score_)
print('GridSearch params:', gs.best_params_)
```

```python
params = {
    'n_estimators' : np.arange(5, 200, 5)
}
gs = GridSearchCV(RandomForestRegressor(), param_grid= params)
gs.fit(numeric_train, target)
print('Gradient Boost GridSearch best score:', gs.best_score_)
print('GridSearch params:', gs.best_params_)

```

```python
pred = gs.predict(numeric_test)
pred.shape
```

```python
test['SalePrice'] = pred
```

```python
test[['Id', 'SalePrice']].to_csv('submission.csv', index = False) 

# Submitted 1st draft to Kaggle. Score: 37094.436
# Let's see if I can optimize it.
```

### Iteration 2: GridSearch Submission w/ Standard Scale (submission2.csv)

```python
# Let's resubmit with scaled values...

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()

scaled_numeric_train = ss.fit_transform(numeric_train)
X = scaled_numeric_train
y = target
```

```python
X = pd.DataFrame(X)
X.columns = numeric_train.columns
X.shape
```

```python
params2 = {
    'n_estimators' : np.arange(5, 200, 5)
}
gs = GridSearchCV(RandomForestRegressor(), param_grid= params2)
gs.fit(X, y)
print('RandomForest GridSearch best score:', gs.best_score_)
print('GridSearch params:', gs.best_params_)
```

```python
params2 = {
    'n_estimators' : np.arange(5, 200, 5)
}
gs = GridSearchCV(GradientBoostingRegressor(), param_grid= params2)
gs.fit(X, y)
print('Gradient Boost GridSearch best score:', gs.best_score_)
print('GridSearch params:', gs.best_params_)
```

```python
pred = gs.predict(numeric_test)
```

```python
test['SalePrice'] = pred
```

```python
test[['Id', 'SalePrice']].to_csv('submission2.csv', index = False)

# Submission got scored: 37094.436 no change.
# It's interesting the scaled values made no difference. Let's optimize it further by taking out some multicollinear features.
```

### Iteration 3: Gridsearch Submission w/ Standard Scale and one pair of Multicollinearity columns taken out (submission3.csv)


Since I imputed -999 for missing values for RandomForest and GB models, I have to impute missing values differently for this method.

```python
og_numeric_train["Mas Vnr Area"].fillna(og_numeric_train["Mas Vnr Area"].mean(), inplace=True)
og_numeric_train['Lot Frontage'].fillna(og_numeric_train["Lot Frontage"].mean(), inplace=True)
```

```python
og_numeric_train = og_numeric_train.dropna()
```

```python
# First we need to see what columns are worth taking out by using a corr. heatmap.

plt.subplots(figsize = (25, 25))
corr = numeric_train.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

with sns.axes_style('white'):
    ax = sns.heatmap(corr, mask=mask, square = True, annot = True)

plt.show()
```

### Pairs that are multicollinear:
- **Year Built** and Year Remod/Add
- **Gr Liv Area** and **2nd Flr SF** (Took out Gr Liv Area as well because it hits a lot of other columns)

```python
multi_coll = ['Year Built', 'Gr Liv Area', '2nd Flr SF']
# Identified columns I want to take out that demonstrated high multicollinearity.

for i in multi_coll:
    del numeric_train[i] 
```

```python
# Took the same columns out for test df.
for i in multi_coll:
    del numeric_test[i]
```

```python
numeric_train.columns == numeric_test.columns # they line up!
```

```python
# After we took out multicollinear columns, let's try it again.

params3 = {
    'n_estimators' : np.arange(5, 200, 5)
}
gs = GridSearchCV(RandomForestRegressor(), param_grid= params3)
gs.fit(numeric_train, y)
print('Random Forest GridSearch best score:', gs.best_score_)
print('GridSearch params:', gs.best_params_)

```

```python
# After we took out multicollinear columns, let's try it again.

params3 = {
    'n_estimators' : np.arange(5, 200, 5)
}
gs = GridSearchCV(GradientBoostingRegressor(), param_grid= params3)
gs.fit(numeric_train, y)
print('Gradient Boost GridSearch best score:', gs.best_score_)
print('GridSearch params:', gs.best_params_)

```

```python
pred = gs.predict(numeric_test)
```

```python
test['SalePrice'] = pred
test[['Id', 'SalePrice']].to_csv('submission3.csv', index = False) # This scored 31359.65063! Best so far.
```

Taking multicollinear features doesn't improve the score... hmm...

### Iteration 4: Feature Selection using LassoCV and SelectFromModel (submission4.csv)

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LassoCV

# We use the base estimator LassoCV since the L1 norm promotes sparsity of features.
clf = LassoCV()

# Set a minimum threshold of 0.25
sfm = SelectFromModel(clf, threshold = 0.25)
sfm.fit(X, y)
n_features = sfm.transform(X).shape[1]


# Reset the threshold till the number of features equals two.
# Note that the attribute can be set directly instead of repeatedly
# fitting the metatransformer.
while n_features > 5:
    sfm.threshold += 0.1
    X_transform = sfm.transform(X)
    n_features = X_transform.shape[1]
    
```

```python
X.columns[sfm.get_support()]
```

```python
# Now that I have the features I want selected, let's run it again!
lasso_selection_from_model = ['Overall Qual', 'Year Built', 'BsmtFin SF 1', 'Gr Liv Area', 'Garage Area']
```

```python
params4 = {
    'n_estimators' : np.arange(5, 200, 5)
}
gs = GridSearchCV(RandomForestRegressor(), param_grid= params4)
gs.fit(numeric_train[lasso_selection_from_model], y)
print('GridSearch best score:', gs.best_score_)
print('GridSearch params:', gs.best_params_)
```

```python
params4 = {
    'n_estimators' : np.arange(5, 200, 5)
}
gs = GridSearchCV(GradientBoostingRegressor(), param_grid= params4)
gs.fit(numeric_train[lasso_selection_from_model], y)
print('GridSearch best score:', gs.best_score_)
print('GridSearch params:', gs.best_params_)

```

Model built with best features do not score better.
